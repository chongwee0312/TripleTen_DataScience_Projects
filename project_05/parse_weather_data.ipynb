```python
# Import libraries
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# get the Chicago weather information in November 2017 from the website
URL = 'https://practicum-content.s3.us-west-1.amazonaws.com/data-analyst-eng/moved_chicago_weather_2017.html'
req = requests.get(URL)
```

```python
# Create a BeautifulSoup instant to manipulate the weather data
soup = BeautifulSoup(req.text, 'lxml')

# Have a look at the weather data
print(soup.prettify())
```

```python
# Find the table containing weather data
table = soup.find('table', attrs={'id': 'weather_records'})
```

```python
# Store the table header in a list
table_header = []

for row in table.find_all('th'):
    table_header.append(row.text)

# Have a look at the table header
table_header
```

```python
# Store the table content in a list
table_content = []

for row in table.find_all('tr'):
    if not row.find_all('th'):
        table_content.append([element.text for element in row.find_all('td')])

# Have a look at the table content
table_content
```

```python
# Create a dataframe containing the weather data
weather_data = pd.DataFrame(table_content, columns = table_header)

# Check the first 5 rows of the weather_data
weather_data.head()
```

```python
# Export the weather data as a csv file
weather_data.to_csv('chicago_weather_nov_2017.csv')
```
